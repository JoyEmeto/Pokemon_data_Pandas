### Regex Filtering (filter based on textual patterns)

```python
#Filter out words using regular expression
#df.loc[df['Type 1'].str.contains("Fire|Grass", regex=True)]
#df.loc[df['Type 1'].str.contains("fire|grass", regex=True)] #this is lover case and would only out put the words that are in lower case

import re
df.loc[df['Type 1'].str.contains("fire|grass", flags=re.I, regex=True)]
#flags = re.I is used to ignore the word case
#regex=True shows that it has a regular expression (|)
```
